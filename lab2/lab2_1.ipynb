Программа по введённому аргументу x выводит значение функции y=f(x) при этом аргументе:

```python
x = float(input("Введите значение аргумента x: "))
```

```python
print(f"y1 = {x}")

if x >= 5:
    y =  -(x + 5)**2 + 9
    print(f"y2 = {y:g}")
else: 
    print("x не определён для функции 2")

print(f"f3 = {-(x + 5)**2 + 9}")

if -2 <= x <= 0: 
    y = -((x - 5)**0.5) + 5 
    print(f"{y:g}")
else: 
    print("x не определён для функции 4")
```

```python
x, y = map(float, input("Введите через пробел значения x,y: ").split()) #Ввод значений x,y
```

```python
lower_than_red = (y <= 2*x - 4) #Принимает True, если лежит ниже чем красная линия
higher_than_blue = (y >= x/4 - 1/2) #Принимает True, если лежит выше чем синия линия
lower_than_green = y <= -(x - 4)**2 + 4 #Принимает True, если лежит ниже чем зеленая линия
higher_than_purpe = y >= -0.75*x - 4 #Принимает True, если лежит выше чем фиолетовая линия

if lower_than_red and higher_than_blue and lower_than_green and x > 2: #Проверяем условия нахождения в правой области
    print("Точка лежит в правой области")

elif x == 2 and y == 0: #Проверяем лежит ли точка на пересечении двух областей
    print("Точка лежит на пересечении двух областей")

elif not lower_than_red and not higher_than_blue and higher_than_purpe and x < 2: #Проверяем условия нахождения в левой области
    print("Точка лежит в левой области")
    
else: #Если точка не подошла ни под одно условие, значит она не лежит внутри этих областей
    print("Точка не лежит ни в какой области")
```
