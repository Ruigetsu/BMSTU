```python
# y = -x**2 - 5

first_arg = float(input("Введите начальное знач. аргумента: "))
last_arg = float(input("Введите конечное знач. аргумента: "))
step = float(input("Введите шаг: "))
```

```python
cur_x = first_arg
last_x = last_arg
y_min = -cur_x**2 - 5
y_max = -cur_x**2 - 5
while cur_x <= last_x:
    y = -cur_x**2 - 5
    y_min = min(y_min,y)
    y_max = max(y_max,y)
    cur_x = round(cur_x+step,4)

width = 100

if len(str(first_arg)) > 5:
    offset = len(str(first_arg))
else: 
    offset = 5
row_of_ticks = " "*offset

val_per_symbol = (y_max - y_min)/width

pos_zero = 0

pos_min_tick = 1
pos_max_tick = int((y_max - y_min)//val_per_symbol) - len(str(y_max))
flag_of_0 = 0
if y_min < 0 < y_max: 
    pos_zero = int((0 - y_min)//val_per_symbol)
    row_of_ticks += f'{" "*pos_min_tick}{y_min:.2g}'
    row_of_ticks += f'{" "*(pos_zero-pos_min_tick)}0'
    row_of_ticks += f'{" "*pos_max_tick-pos_zero}{y_max:.2g}'
elif 0 < y_min:
    pos_zero = 0
    row_of_ticks += f'{" "*pos_min_tick}0{y_min:.2g}'
    row_of_ticks += f'{" "*pos_max_tick}{y_max:.2g}'
    flag_of_0 = 1
elif y_max < 0:
    pos_zero = pos_max_tick + len(f"{y_max:.2g}") + 3
    row_of_ticks += f'{" "*pos_min_tick}{y_min:.2g}'
    row_of_ticks += f'{" "*pos_max_tick}{y_max:.2g}'
    row_of_ticks += f'{" "*(3)}0'
    flag_of_0 = 2
print(row_of_ticks)

x = first_arg
while x <= last_arg:
    y = -x**2 - 5
    point_pos = int((y - y_min)//val_per_symbol)
    if flag_of_0 == 0:
        if point_pos < pos_zero:
            print(f'{x:>{offset - 1}.2g}|{' ' * point_pos}*{' ' * (pos_zero - point_pos - 1)}|')
        elif point_pos > pos_zero:
            print(f'{x:>{offset - 1}.2g}|{' ' * pos_zero}|{' ' * (point_pos - pos_zero - 1)}*')
        else: 
            print(f'{x:>{offset - 1}.2g}|{' ' * point_pos}+')
    elif flag_of_0 == 1:
        print(f'{x:>{offset - 1}.2g}|{' ' * point_pos}*')
    else:
        if x == 1.5:
            print(point_pos,pos_zero) 
        print(f'{x:>{offset - 1}.2g}|{' ' * point_pos}*{" "*(len(row_of_ticks) - point_pos - offset - 2)}|')
        
    x = round(x+step,4)



```
